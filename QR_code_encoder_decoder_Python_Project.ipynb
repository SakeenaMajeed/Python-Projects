```python
# !pip install qrcode opencv-python pillow
import qrcode
from PIL import Image

def generate_qr(data, filename="qrcode.png"):
    qr = qrcode.QRCode(
        version=1,
        error_correction=qrcode.constants.ERROR_CORRECT_L,
        box_size=10,
        border=4,
    )
    qr.add_data(data)
    qr.make(fit=True)

    img = qr.make_image(fill="black", back_color="white")
    img.save(filename)
    display(img)

generate_qr("https://www.linkedin.com/in/sakeena-majeed-86b58732a/", "sakeena_qr.png")

```
