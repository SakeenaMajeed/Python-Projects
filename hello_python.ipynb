```python
import streamlit as st
from PIL import Image
import time


st.set_page_config(page_title="AI-Powered Web App", page_icon="ü§ñ", layout="wide")

st.markdown("""
    <style>
        body {
            background-color: #f9f9f9;
            font-family: 'Poppins', sans-serif;
        }
        .main-title {
            text-align: center;
            font-size: 48px;
            font-weight: bold;
            color: #2E86C1;
            animation: fadeIn 2s ease-in-out;
        }
        .stButton>button {
            background: linear-gradient(90deg, #4CAF50, #2E86C1);
            color: white;
            font-size: 18px;
            padding: 12px;
            border-radius: 12px;
            transition: 0.3s ease;
        }
        .stButton>button:hover {
            transform: scale(1.05);
        }
        .stTextInput>div>div>input, .stTextArea>div>textarea {
            font-size: 16px;
            border-radius: 8px;
            border: 1px solid #2E86C1;
            padding: 8px;
        }
        @keyframes fadeIn {
            from { opacity: 0; }
            to { opacity: 1; }
        }
    </style>
""", unsafe_allow_html=True)

if 'quiz_started' not in st.session_state:
    st.session_state.quiz_started = False
if 'quiz_submitted' not in st.session_state:
    st.session_state.quiz_submitted = False
if 'quiz_score' not in st.session_state:
    st.session_state.quiz_score = 0
if 'leaderboard' not in st.session_state:
    st.session_state.leaderboard = []


image_path = "Python-01.jpg"
try:
    image = Image.open(image_path)
    st.image(image, caption="Technology at its best", use_container_width=True)
except FileNotFoundError:
    st.error("Image not found! Please upload a valid image.")


st.markdown("## Welcome to Python Website")


python_code = st.text_area("Write Python Code Here:")
if st.button("Submit Code"):
    with st.spinner("Executing..."):
        time.sleep(2)
    st.balloons()
    st.success("Code Executed Successfully! üéâ")

user_name = st.text_input("Enter Your Name to Start Quiz:")
if user_name:
    if not st.session_state.quiz_started:
        if st.button("Start Quiz"):
            st.session_state.quiz_started = True
            st.session_state.quiz_submitted = False
            st.session_state.quiz_score = 0

    if st.session_state.quiz_started:
        st.write(f"Hello, {user_name}! Let's start the quiz.")
        progress_bar = st.progress(0)


        question_1 = st.radio("1. What is the output of `print(2 * 3)`?", ["5", "6", "8"], key="q1")
        progress_bar.progress(20)


        question_2 = st.radio("2. Which data type is immutable in Python?", ["List", "Dictionary", "Tuple"], key="q2")
        progress_bar.progress(40)


        question_3 = st.radio("3. What is the result of `10 // 3`?", ["3.33", "3", "4"], key="q3")
        progress_bar.progress(60)


        question_4 = st.radio("4. Which keyword is used to define a function in Python?", ["func", "def", "function"], key="q4")
        progress_bar.progress(80)


        question_5 = st.radio("5. What is the result of `3 ** 2`?", ["6", "9", "12"], key="q5")
        progress_bar.progress(100)

        if st.button("Submit Quiz"):
            st.session_state.quiz_submitted = True
            if question_1 == "6":
                st.session_state.quiz_score += 1
            if question_2 == "Tuple":
                st.session_state.quiz_score += 1
            if question_3 == "3":
                st.session_state.quiz_score += 1
            if question_4 == "def":
                st.session_state.quiz_score += 1
            if question_5 == "9":
                st.session_state.quiz_score += 1

        if st.session_state.quiz_submitted:
            st.write(f"Your Quiz Score: {st.session_state.quiz_score}/5")
            if st.session_state.quiz_score == 5:
                st.balloons()
                st.success(f"Congratulations, {user_name}! You scored perfectly! üéâüéà")


            st.session_state.leaderboard.append((user_name, st.session_state.quiz_score))

            st.markdown("### üèÜ Leaderboard")
            leaderboard_sorted = sorted(st.session_state.leaderboard, key=lambda x: x[1], reverse=True)
            st.table(leaderboard_sorted)


st.markdown("""
    <hr>
    <p style='text-align: center; font-size: 14px;'>Generate By ‚ù§Ô∏è Sakeena Majeed</p>
""", unsafe_allow_html=True)
```
